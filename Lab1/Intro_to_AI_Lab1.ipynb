练习3和练习2的目的相同，使用print打印出所有可能的路线，要求经过5个港口中的每一个。在练习3中，我们使用更加高级的算法——递归。

首先以空列表开始递归，命名该列表为`route`。递归函数应该涉及到一个for循环，该循环遍历可用的港口，命名为`ports`，并为每个港口：

1. 将其添加到`route`列表
2. 将其从可用的港口列表`ports`中删除，并且
3. 递归调用剩余的港口
    
当递归函数被调用时，如果港口列表`ports`为空，递归应该停止，并打印出`route`列表。

例如，如果我们以 `route = [0]`和 `ports = [1, 2, 3, 4]` 开始递归，意味着第一站总是巴拿马，首先要做的是尝试 route 中 1 之后的每个港口，得到例如 `route = [0, 3]`并继续递归其余的端口，例如` [1, 2, 4]`。要从列表中删除第` i` 项，可以写成` ports[:i]+ports[i+1:]`。

```python
portnames = ["PAN", "AMS", "CAS", "NYC", "HEL"]
 
def permutations(route, ports):
    if not ports:
        # Print the current route
        print(' '.join([portnames[i] for i in route]))
    else:
        for i in range(len(ports)):
            # 将第i个港口添加到route列表
            None
            
            # 将第i个港口从可用的港口列表ports中删除，剩余可用港口列表命名为rest
            None
            
            # 递归调用剩余的港口
            None        
            
            # 从route中删除当前端口，以便我们可以尝试下一个端口
            route.pop()


# 这将以0（"SH"）为第一站开始递归。
permutations([0], list(range(1, len(portnames))))
```
